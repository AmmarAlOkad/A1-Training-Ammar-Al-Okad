## What is Multiple Inheritance?

In Python, multiple inheritance means a class can inherit from more than one parent class.

A child class can combine features from multiple classes at the same time. This helps reuse code and combine behaviors.


## Simple Example

Letâ€™s imagine a child learning different things from both parents.


```python
class Father:
    def skill(self):
        print("Father knows gardening")

class Mother:
    def talent(self):
        print("Mother knows painting")

class Child(Father, Mother):
    def hobby(self):
        print("Child loves football")

c = Child()

c.skill()
c.talent()
c.hobby()
```

## How Python Chooses Between Methods

If two parent classes have methods with the same name, Python follows something called Method Resolution Order (MRO).

It checks classes in a specific order.


```python
print(Child.__mro__)
```

## Summary

- A class can inherit from multiple classes.
- It allows combining different behaviors.
- Python uses MRO to decide which method to run.
- It is powerful, but should be used carefully.
