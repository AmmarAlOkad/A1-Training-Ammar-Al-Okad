#  Built-in Packages and Modules in Python

## What are Built-in Modules?

Built-in modules are ready-made Python files that come with Python.
They contain useful functions and tools so we don’t have to write everything from scratch.

We use the import keyword to use them.


## 1️ math Module

The math module helps us perform mathematical operations.


```python
import math

print(math.sqrt(25))
print(math.pi)
```

## 2️ random Module

The random module helps us generate random numbers.


```python
import random

print(random.randint(1, 10))
```

## 3️ datetime Module

The datetime module helps us work with dates and time.


```python
import datetime

today = datetime.date.today()
print(today)
```

## os modual

```python
import os

print(os.getcwd()) 
print(os.listdir())    
```

## sys Module

```python
import sys

print(sys.version)
print(sys.platform)
```

## statistics Module

```python
import statistics

numbers = [10, 20, 30, 40]

print(statistics.mean(numbers)) 
print(statistics.median(numbers))
```

## time Module

```python
import time

print(time.time())
time.sleep(2) 
print("Done")
```

## What is a Package?

A package is a collection of modules inside a folder.
For example, `datetime` is a module, and Python’s standard library contains many modules grouped together.

Packages help organize related modules.


## Summary

- Built-in modules come with Python.
- We import them using the `import` keyword.
- Examples include math, random, and datetime.
- Packages help organize multiple modules together.
