```python
import pandas as pd
print()

original_df = pd.read_csv("dirty_cafe_sales.csv")
df = original_df.drop_duplicates()
df
```

Quantity

```python

Quantity_Mode_Val = df["Quantity"].mode()[0]
print("Quantity_Mode_Val\n", Quantity_Mode_Val)
df["Quantity"] = df["Quantity"].str.strip()
df["Quantity"] = df["Quantity"].fillna(Quantity_Mode_Val)
# df["Item"] = df["Item"].str.replace('', Quantity_Mode_Val)
df["Quantity"] = df["Quantity"].str.replace('ERROR', Quantity_Mode_Val)
df["Quantity"] = df["Quantity"].str.replace('UNKNOWN', Quantity_Mode_Val)
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df.info()
```

Price Per Unit  &  Total Spent

```python

df["Price Per Unit"] = df["Price Per Unit"].str.strip()
df["Price Per Unit"] = df["Price Per Unit"].fillna('')
df["Price Per Unit"] = df["Price Per Unit"].str.replace('ERROR', '')
df["Price Per Unit"] = df["Price Per Unit"].str.replace('UNKNOWN', '')
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"])
###################
df["Total Spent"] = df["Total Spent"].str.strip()
df["Total Spent"] = df["Total Spent"].fillna('')
df["Total Spent"] = df["Total Spent"].str.replace('ERROR', '')
df["Total Spent"] = df["Total Spent"].str.replace('UNKNOWN', '')
df["Total Spent"] = pd.to_numeric(df["Total Spent"])

groupby_data = df.groupby("Item")["Price Per Unit"]
print(groupby_data.max(), "\n")
# print(groupby_data.min(), "\n")

for x in df.index:
    if pd.isna(df.loc[x, "Price Per Unit"]):
        if df.loc[x, "Item"] == "Cake":
            df.loc[x, "Price Per Unit"] = 3.0
        elif df.loc[x, "Item"] == "Coffee":
            df.loc[x, "Price Per Unit"] = 2.0
        elif df.loc[x, "Item"] == "Cookie":
            df.loc[x, "Price Per Unit"] = 1.0
        elif df.loc[x, "Item"] == "Juice":
            df.loc[x, "Price Per Unit"] = 3.0
        elif df.loc[x, "Item"] == "Salad":
            df.loc[x, "Price Per Unit"] = 5.0
        elif df.loc[x, "Item"] == "Sandwich":
            df.loc[x, "Price Per Unit"] = 4.0
        elif df.loc[x, "Item"] == "Smoothie":
            df.loc[x, "Price Per Unit"] = 4.0
        elif df.loc[x, "Item"] == "Tea":
            df.loc[x, "Price Per Unit"] = 1.5
        
###################

# for x in df.index:
#     if pd.isna(df.loc[x, "Total Spent"]):
#         df.loc[x, "Total Spent"] = df.loc[x, "Price Per Unit"] * df.loc[x, "Quantity"]

df["Total Spent"] = df["Total Spent"].fillna( df["Price Per Unit"] * df["Quantity"] )
df["Price Per Unit"] = df["Price Per Unit"].fillna( df["Total Spent"] / df["Quantity"] )

df.info()
```

Item

```python

df["Item"] = df["Item"].str.strip()
df["Item"] = df["Item"].fillna("")
df["Item"] = df["Item"].str.replace('ERROR', "")
df["Item"] = df["Item"].str.replace('UNKNOWN', "")

for x in df.index:
    # if pd.isna(df.loc[x, "Item"]):
    if df.loc[x, "Item"] == "":
        if df.loc[x, "Price Per Unit"] == 3.0:
            df.loc[x, "Item"] = "Cake"
        elif df.loc[x, "Price Per Unit"] == 2.0:
            df.loc[x, "Item"] = "Coffee"
        elif df.loc[x, "Price Per Unit"] == 1.0:
            df.loc[x, "Item"] = "Cookie"
        elif df.loc[x, "Price Per Unit"] == 3.0:
            df.loc[x, "Item"] = "Juice"
        elif df.loc[x, "Price Per Unit"] == 5.0:
            df.loc[x, "Item"] = "Salad"
        elif df.loc[x, "Price Per Unit"] == 4.0:
            df.loc[x, "Item"] = "Sandwich"
        elif df.loc[x, "Price Per Unit"] == 4.0:
            df.loc[x, "Item"] = "Smoothie"
        elif df.loc[x, "Price Per Unit"] == 1.5:
            df.loc[x, "Item"] = "Tea"

     
```

Payment Method

```python
Payment_Mode_Val = df["Payment Method"].mode()[0]
print("Payment_Mode_Val\n", Payment_Mode_Val)

df["Payment Method"] = df["Payment Method"].str.strip()
df["Payment Method"] = df["Payment Method"].fillna(Payment_Mode_Val)
df["Payment Method"] = df["Payment Method"].str.replace('ERROR', Payment_Mode_Val)
df["Payment Method"] = df["Payment Method"].str.replace('UNKNOWN', Payment_Mode_Val)
df.info()
```

Location

```python

Location_Mode_Val = df["Location"].mode()[0]
print("Location_Mode_Val\n", Location_Mode_Val)

df["Location"] = df["Location"].str.strip()
df["Location"] = df["Location"].fillna(Location_Mode_Val)
df["Location"] = df["Location"].str.replace('ERROR', Location_Mode_Val)
df["Location"] = df["Location"].str.replace('UNKNOWN', Location_Mode_Val)
df.info()
```

Transaction Date

```python

df["Transaction Date"] = df["Transaction Date"].str.strip()
df["Transaction Date"] = df["Transaction Date"].fillna('')
df["Transaction Date"] = df["Transaction Date"].str.replace('ERROR', '')
df["Transaction Date"] = df["Transaction Date"].str.replace('UNKNOWN', '')
######################

df["Transaction Date"] = pd.to_datetime(df["Transaction Date"])

df.info()
```

Season

```python
seasons_map = {
    12: "Winter", 1: "Winter", 2: "Winter",
    3: "Spring", 4: "Spring", 5: "Spring",
    6: "Summer", 7: "Summer", 8: "Summer",
    9: "Fall", 10: "Fall", 11: "Fall"
}

df["Season"] = df["Transaction Date"].dt.month.map(seasons_map)
```
